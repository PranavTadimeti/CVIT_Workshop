```
# counting boundary coins

import skimage.morphology as morph
i = cv2.imread('/content/coins.jpg')
i = cv2.cvtColor(i, cv2.COLOR_BGR2GRAY)
im = np.uint8(i)
kernel = morph.disk(10)
o = cv2.erode(im,kernel)
o[10:o.shape[0] - 10,10:o.shape[1]-10] = 0
# out = cv2.morphologyEx(im, cv2.MORPH_HITMISS, kernel)
o = cv2.dilate(o,kernel)
cv2_imshow(o)
```

```
i = cv2.imread('/content/coins.jpg')
i = cv2.cvtColor(i, cv2.COLOR_BGR2GRAY)
im = np.uint8(i)
kernel = morph.disk(10)
o = cv2.erode(im,kernel)

last = 0
for i in range(len(o)):
  for j in range(len(o[i])):
    if(o[i][j] != 0):
      if(j - last > 9):
        o[i][j] = 0
        o [i][last] = 0
      last = j
last = 0

for j in range(len(o[0])):
  for i in range(len(o)):
    if(o[i][j] != 0):
      if(j - last > 9):
        o[i][j] = 0
        o[i][last] = 0
      last = j

o = cv2.dilate(o,kernel)    

cv2_imshow(o)
```
