```python
# finding distance of line from origin

from math import *

def dist_from_origin(a,b,c):
    dist= abs(c) / (sqrt(a^2 + b^2))
    
    return dist

d1 = dist_from_origin(1,0,-50)
d2 = dist_from_origin(0,1,-25)
print(d1,d2)
```

```python
# plane plot

from mpl_toolkits.mplot3d import Axes3D as ax
%matplotlib inline

a = 1
b = 1
c = 0
d = 10

y = np.linspace(-5,5,100)
z = np.linspace(-5,5,100)

Y,Z = np.meshgrid(y,z)
X = (-b*Y-c*Z-d)/a

p = plt.figure().gca(projection='3d')
p.plot_surface(X,Y,Z)

```

```python
# gaussian surface 3D plot

x=np.linspace(-10,10, num=100)
y=np.linspace(-10,10, num=100)

z=np.exp(-0.1*x**2-0.1*y**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x,y,z)
```
