```python
# inverse of matrix

import numpy as np

M = np.random.randn(3,3)
U,S,V = np.linalg.svd(M)

M_inv = np.transpose(V) @ np.diag(1/S) @ np.transpose(U)

print(M)
print(M_inv)
```
